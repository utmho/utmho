```python
df["target"] = df.shift(-1)["t_max"]
df = df.iloc[:-1,:].copy()
df
```

```python
from sklearn.linear_model import Ridge
reg = Ridge(alpha=.1)
```

```python
predictors = df.columns[~df.columns.isin(["target"])]  # all of column except "target"
### or predictors = ["rain", "t_min", "t_max","rh_min", "rh_max", "ff_max_wind",	"dd_max_wind"]
```

```python
def backtest(df, model, predictors, start=3650, step=90): # start day 3650 is about 10 years of initail data and step 90 days
    all_predictions = []

    for i in range(start, df.shape[0], step):
        train = df.iloc[:i,:]
        test = df.iloc[i:(i+step),:]

        model.fit(train[predictors], train["target"])

        preds = model.predict(test[predictors])
        preds = pd.Series(preds, index=test.index)
        combined = pd.concat([test["target"], preds], axis=1)
        combined.columns = ["actual", "prediction"]
        combined["diff"] = (combined["prediction"] - combined["actual"]).abs()

        all_predictions.append(combined)
    return pd.concat(all_predictions)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_predictions(predictions):
    plt.figure(figsize=(10, 6))
    predictions['actual'].plot(label='Actual')
    predictions['prediction'].plot(label='Prediction')
    plt.xlabel('Index')
    plt.ylabel('Value')
    plt.title('Actual vs. Prediction')
    plt.legend()
    plt.show()

# Call the backtest function and store the result in a variable
result = backtest(df, model, predictors)

# Call the plot_predictions function with the result
plot_predictions(result)
```

```python
from sklearn.linear_model import Ridge

rr = Ridge(alpha=.1)
```

```python
predictions = backtest(df, rr, predictors)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
mean_absolute_error(predictions["actual"], predictions["prediction"])
```
